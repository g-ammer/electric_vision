```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import os
import scipy.io
```

```python
# Import data from mat-file

data = scipy.io.loadmat('Vtrace_MEC_PTX_Ctrl.mat')
Vtrace_all_Ctrl = data['Vtrace']
data = scipy.io.loadmat('Vtrace_MEC_PTX.mat')
Vtrace_all_Block = data['Vtrace']
```

```python
fs = 10000.0  # define data acquisition rate in Hz
time = np.arange(0, 2800000/fs, 1/float(fs))
```

```python
fig = plt.figure(figsize = (14,2))

ax = fig.add_subplot(121) #PLOT mean Ftraces

z = 0
n = [0,1] 
for i in n:
    plt.plot(time, Vtrace_all_Ctrl[:,i]+z*10, linewidth='0.5',color='0.6')
    z +=1
    
ax = fig.add_subplot(122) #PLOT mean Ftraces
z = 0
n = [0,1] 
for i in n:
    plt.plot(time, Vtrace_all_Block[:,i]+z*10, linewidth='0.5',color='C0')
    z +=1    
    
#plt.savefig('Traces_ephys_Ctrl.pdf',bbox_inches='tight', format='pdf')
plt.show()
```

```python
fig = plt.figure(figsize = (7,3))
#plt.subplots_adjust(wspace=0.3)

ax = fig.add_subplot(121) #PLOT mean Ftraces

plt.plot(time[0:10000], Vtrace_all_Ctrl[1580000:1590000,0], linewidth='1',color='0.4')
plt.plot(time[0:10000], Vtrace_all_Ctrl[8000:18000,1]+8, linewidth='1',color='0.4')
    
ax.set_ylim([-3.5,14])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)


ax = fig.add_subplot(122) #PLOT mean Ftraces
z = 0
n = [0,1] 
for i in n:
    plt.plot(time[0:10000], Vtrace_all_Block[9000:19000,i]+z*8, linewidth='1',color='C0')
    z +=1
ax.set_ylim([-3.5,14])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.savefig('VTraces_MEC_PTX_zoom.pdf',bbox_inches='tight', dpi=600, transparent=True)
plt.show()
```
